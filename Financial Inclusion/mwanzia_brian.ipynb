**Bivariate Analysis**


Bivariate analysis is the simultaneous analysis of two variables (attributes). It explores the concept of relationship between two variables, whether there exists an association and the strength of this association, or whether there are differences between two variables and the significance of these differences.

**1.People in rural areas have a low chance to have bank accounts..**



```python
sns.catplot(data=data, x='location_type', kind='count', hue='bank_account')
```

From the above plot, we can see that a majority of the people living in rural areas dont have bank accounts. This can be due to a number of factors such as:


1.   They do not trust the bank to keep their money safe.
2.   Bank services in rural are unreliable.
3.   The withdrawal and other transaction fees are prohibitively expensive.



**2. People who have mobile phones have alower chance to use bank accounts because of mobile money services.**

```python
plt.figure(figsize=(16, 4))
sns.catplot(data=data, x='cellphone_access', kind='count', hue='bank_account').set(title='Cellphone Access')
plt.xticks(ha = 'right')
```

From the cellphone Access plot we can see that most of the people that have access to cellphone do not have bank accounts. This provess that people who have mobile phones do not have bank accounts. This is usually due to:


1.   Mobile money services make money transfer faster and more convinient over a wide territorial reach.
2.   mobile money services which is more accessible and affordable especially for people living in rural areas.



**3. Females have less chance to have bank account.**

```python
plt.figure(figsize=(16, 4))
sns.catplot(data=data, x='gender_of_respondent', kind='count', hue='bank_account').set(title='Gender_of_respondent vs Bank_account')
```

The plot shows that there is a small difference between males and females who have bank accounts. The number of males are greater than females.



```python
plt.figure(figsize=(16, 6))
sns.catplot(data=data, x='gender_of_respondent', kind='count', hue='cellphone_access').set(title='Gender vs cellphone_access')
plt.xticks(rotation = 45,
           ha = 'right')
```

From the Gender vs cellphone_access plot we can see that more women have phones as compared to men.

This provides them with a higher chance of using mobile phone-based money transfer service, rather than having access to bank accounts.

**4. People who are employed have a higher chance of having a bank account than people who are unemployed.**

```python
plt.figure(figsize=(16, 4))
sns.catplot(data=data, x='job_type', kind ='count', hue='bank_account').set(title = 'Job Type')
plt.xticks(
    rotation = 45,
    ha = 'right'
)
```

The job_type plot shows that formally employed Private and formally employed Government have access to bank accounts.

It also shows that the majority of people who are self_employed don't have access to bank accounts followed by informally employed then farming and finally fishing.

**5. People with low education have low chance to have bank accounts.**

```python
plt.figure(figsize=(16, 4))
sns.catplot(data=data, x='education_level', kind ='count', hue='bank_account').set(title = 'Education Level')
plt.xticks(
    rotation = 45,
    ha = 'right'
)
```

The education level plot show that majority of people with only a primary education or lower have a higher chance of not having a bank account.

This satisfies our assumptions that people with a lower education level have a lower chance of having a bank account.


**6. Does ones position in the family influence their access to a bank account.**

```python
plt.figure(figsize=(16, 4))
sns.catplot(data=data, x='relationship_with_head', kind ='count', hue='bank_account').set(title = 'Relatioship_with_family_head')
plt.xticks(
    rotation = 45,
    ha = 'right'
)
```
